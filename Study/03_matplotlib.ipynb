# matplotlib
The module of Python for plotting is **Matplotlib**.

Matplotlib is useful for creating a wide range of plots, including line plots, scatter plots, bar plots, histograms, and more. It is particularly beneficial for visualizing data in the form of matrices and arrays, making it an essential tool for data analysis and scientific computing.

## 2D Graphs <a class="anchor" id="first"></a>

### 1. Line Plotting
Use 'plt.plot()' to create simple 2D line plots.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

### 2. Scatter Plotting
Use 'plt.scatter()' to create 2D scatter plots.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y, color='b', label='Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')
plt.legend()
plt.show()
```

### 3. Bar Plotting: 
Use plt.bar() to create vertical or horizontal bar plots.

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [4, 7, 2, 6, 4]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot')
plt.show()
```

### 4. Many graphs

The `subplots()` function is used to create subplots within a single figure.

**Parameters:**

- nrows (int): The number of rows of subplots.
- ncols (int): The number of columns of subplots.

**Example:**

```python
# Create data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1)

# Plot data on the subplots
ax1.plot(x, y1)
ax1.set_title('Subplot 1: Sine Function')

ax2.plot(x, y2)
ax2.set_title('Subplot 2: Cosine Function')
plt.rcParams['figure.figsize'] = [3, 3]
plt.tight_layout()
plt.show()
```

### 5. Circles

**Ex.** Implement a program to visualize the Gershgorin circles for a given $n\times n$ complex matrix. See [Gershgorin theorem on Wikipedia](https://en.wikipedia.org/wiki/Gershgorin_circle_theorem)


```python
import matplotlib.pyplot as plt

N = 4

np.random.seed(55555)

D = np.diag([0, 2 + 2j, 1 - 1j, 5 + 5j]) # Creating complex matrix
M = np.random.rand(N, N) + D

R = np.zeros(N) # Initialize Ri at 0
for i in range(N):
    for l in range(N):
        for j in range(N):
            if j!=l:
                R[i] += abs(M[l,j])

eigenvalues = np.linalg.eigvals(M) # Every eigenvalue of A lies inside of at least one of the discs.

# Plotting code
fig, ax = plt.subplots()
for k in range(N):
    x, y = M[k,k].real, M[k,k].imag
    ax.add_artist( plt.Circle((x, y), R[k], alpha=0.5) )
    plt.plot(eigenvalues[k].real, eigenvalues[k].imag, 'k+')

ax.axis([-4, 12.5, -4, 9])
ax.set_aspect(1)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Gershgorin disks and eigenvalues $x + iy$")
plt.grid()
```

## 3D Graphs <a class="anchor" id="second"></a>
### 1. 3D Line Plotting: 


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Creating data points
x = np.linspace(0, 1, 100)
y = np.sin(x * 2 * np.pi)
z = np.cos(x * 2 * np.pi)

# Plotting the 3D line
ax.plot(x, y, z)
plt.show()
```

### 2. 3D Scatter Plotting: 


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generating random data points
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Creating a 3D scatter plot
ax.scatter(x, y, z, c='r', marker='o')
plt.show()
```

### 3. 3D Surface Plotting

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Creating a meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Plotting the 3D surface
ax.plot_surface(x, y, z, cmap='viridis')
plt.show()
```

## Working with Matrices <a class="anchor" id="third"></a>
Matplotlib can be used to visualize matrices and arrays as heatmaps or 2D/3D surface plots. 

You can use functions such as 'plt.imshow()' for displaying 2D matrices as heatmaps and 'axes.plot_surface()' for visualizing 3D arrays as surfaces.


```python
# Create a sample matrix
matrix = np.random.rand(5, 5)
print(matrix)
```

### 2D Heatmaps

```python
# Plotting the matrix as a heatmap
plt.imshow(matrix, cmap='rainbow', interpolation='nearest')
plt.colorbar() #to show the colorbar
plt.show()
```

### 3D Surface Plot

```python
matrix_3d = np.random.rand(5,5)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Creating a meshgrid
x = np.arange(0, 5, 1)
y = np.arange(0, 5, 1)
x, y = np.meshgrid(x, y)

# Plotting the 3D surface
ax.plot_surface(x, y, matrix_3d, cmap='viridis')

plt.show()
```

## Adjust plot properties <a class="anchor" id="fourth"></a>

### Change figure dimensions
You can change the dimensions of the figure using the figsize parameter when creating the figure using plt.figure().

```python
# Changing figure dimensions
fig = plt.figure(figsize=(8, 6))  # Width = 8 inches, Height = 6 inches
```

### Customizing Colors
You can customize colors using the *color* or *c* parameter in various plotting functions. Matplotlib supports various color formats, including color names, hex values, and RGB or RGBA tuples.

```python
# Customizing colors in a line plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10] 

plt.plot(x, y, color='red')  # Using color name
plt.show()
```

### Labels, Titles, Font, Line Style


```python
# Adjusting plot properties
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color='blue', linestyle='--', marker='x')  # Adjusting line style and marker
plt.xlabel('X-axis', fontsize=12)  # Adjusting label font size
plt.ylabel('Y-axis', fontsize=12)
plt.title('Customized Plot', fontsize=14)  # Adjusting title font size
plt.show()

#or
# Access and modify rcParams
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'
plt.rcParams['figure.figsize'] = (8, 6)
```

## Images <a class="anchor" id="fifth"></a>
To display images imported from external files using Matplotlib, you can use the **matplotlib.image** module.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Load an image from an external file
img = mpimg.imread('TarantulaNebula.jpg')

fig = plt.figure(figsize=(5, 5))
# Display the image
plt.imshow(img)
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()
```

## Many images

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some example matrices (replace these with your actual matrices)
matrix1 = np.random.random((10, 10))
matrix2 = np.random.random((10, 10))
matrix3 = np.random.random((10, 10))

# Create a list of matrices
matrices = [matrix1, matrix2, matrix3]

# Set up the plot layout
num_rows = 1
num_cols = len(matrices)

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5))

# Iterate over matrices and display them in subplots
for i in range(num_cols):
    axes[i].imshow(matrices[i], cmap='viridis')
    axes[i].set_title(f'Matrix {i+1}')
    axes[i].axis('off')  # Optional: Turn off axis ticks and labels

# Adjust layout to prevent clipping of titles
plt.tight_layout()

# Show the plot
plt.show()
```
