**Ex. 1** Compute the trace of a square matrix without using the `trace` method.

```python
# @title Solution
import numpy as np

def matrix_trace(A):
    n = A.shape[0]
    # initialization of trace of A
    trace = 0
    for i in range(0, n):
        trace = trace + A[i][i]
    return trace

A = np.array([[1, 3, 4],
              [1, 3, 5],
               [3, 2, 2]])
print("Tr(A) = ", matrix_trace(A))
```

**Ex. 2** Implement the Gram-Schmidt procedure in Python.

```python
# @title Solution
import numpy as np


def gram_schmidt(A):

    (n, m) = A.shape

    for i in range(m):

        q = A[:, i] # i-th column of A

        for j in range(i):
            q = q - np.dot(A[:, j], A[:, i]) * A[:, j]

        if np.array_equal(q, np.zeros(q.shape)):
            raise np.linalg.LinAlgError("The column vectors are not linearly independent") # CHECK!!!!!

        # normalize q
        q = q / np.sqrt(np.dot(q, q))

        # write the vector back in the matrix
        A[:, i] = q

A = np.array([[1, 2, 2],
              [3, 6, 2],
               [-2, -5, 0.5]])

gram_schmidt(A)
print(A)
```

**Ex. 4** Implement the Newton method
$$ x^{(k+1)}=x^{(k)}-\frac{f(x^{(k)})}{f'(x^{(k)})}$$
for finding the zeros of the function $f(x)$. The inputs of the functions must be:
*   `x0`: initial guess
*   `f`: function
*   `df`: derivative of the function f
*   `tol`: tolerance for checking convergence (assume a default value of $10^{-6}$)
*   `maxiter`: maximum number of iterations (assume a default value of $50$)

Test the implementation by solving the equation $x=\cos(x)$ using $x^{(0)}=0$.

```python
# @title Solution
x_ax = []
y_ax = []
def newton(f, df, x0, tol=1e-6, maxiter=50):
    """
    newton(f, Df, x0, tol=1e-6, maxiter=50)
    """
    x = x0
    for k in range(maxiter+1):
        fx = f(x)
        dfx = df(x)
        x_ax.append(k)
        
        dx = fx/dfx # Something to add, we could include a check for dfx!=0
        x = x - dx
        error = abs(dx) # Absolute value of difference between x(k+1) and x(k)
        y_ax.append(error)
        if error <= tol:
            iterations = k
            return (x, error, iterations)
    iterations = k
    return (x, error, iterations)

import math as mt
f = lambda x: x-mt.cos(x)
df = lambda x: 1+mt.sin(x)
x0 = 5
res = newton(f, df, x0)
print(res)
```

**Ex. 5** Implement a program to visualize the Gershgorin circles for a given $n\times n$ complex matrix. See [Gershgorin theorem on Wikipedia](https://en.wikipedia.org/wiki/Gershgorin_circle_theorem)


```python
# @title Solution
import matplotlib.pyplot as plt

N = 4

np.random.seed(55555)

D = np.diag([0, 2 + 2j, 1 - 1j, 5 + 5j]) # Creating complex matrix
M = np.random.rand(N, N) + D

R = np.zeros(N) # Initialize Ri at 0
for i in range(N):
    for l in range(N):
        for j in range(N):
            if j!=l:
                R[i] += abs(M[l,j])

eigenvalues = np.linalg.eigvals(M) # Every eigenvalue of A lies inside of at least one of the discs.

# Plotting code
fig, ax = plt.subplots()
for k in range(N):
    x, y = M[k,k].real, M[k,k].imag
    ax.add_artist( plt.Circle((x, y), R[k], alpha=0.5) )
    plt.plot(eigenvalues[k].real, eigenvalues[k].imag, 'k+')

ax.axis([-4, 12.5, -4, 9])
ax.set_aspect(1)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Gershgorin disks and eigenvalues $x + iy$")
plt.grid()
```

**Ex.6** Solve the ODE $y'=\exp(-t)$ with $y(0)=-1$ using the explicit Euler (EE) method. Implement the method from scratch. Set the final time $T=1$ and $h=0.1$.
The EE method for solving $y'=f(t,y)$ reads
$$\bar{y}_{n+1}=\bar{y}_n+hf_n$$
where $f_n=f(t_n,\bar{y}_n)$.

The exact solution of the problem is $y(t)=-exp(-t)$.

```python
# @title Solution
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
f = lambda t, s: np.exp(-t)
h = 0.1 # Step size
T = 10
t = np.arange(0, T+h , h) # Numerical grid from 0.1 to 10, steps of 0.1
print(t)
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t)) # A vector of zeros
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

**Ex. 7** Implement the conjugate gradient algorithm for solving the linear system $A{\bf x}={\bf b}$. The method is described as follows:


1.   initialize ${\bf x}^{(0)}$.
2.   compute ${\bf r}^{(0)} = {\bf b}-A {\bf x}^{(0)}$
3.   assign ${\bf p}^{(0)}={\bf r}^{(0)}$
4.   for $k=0,1,2,...$ :
    *   calculate $\alpha^{(k)}=\frac{({\bf r}^{(k)})^T{\bf r}^{(k)}}{({\bf p}^{(k)})^TA{\bf p}^{(k)}}$
    *   update ${\bf x}^{(k+1)}={\bf x}^{(k)}+\alpha^{(k)}{\bf p}^{(k)}$
    *   calculate ${\bf r}^{(k+1)} = {\bf b}-A {\bf x}^{(k+1)}= {\bf r}^{(k)}-\alpha^{(k)}A{\bf p}^{(k)}$
    *   
        if $||{\bf r}^{(k+1)}||<\epsilon$ then
            STOP
    * 
        else
            calculate $\beta^{(k)}=\frac{({\bf r}^{(k+1)})^T{\bf r}^{(k+1)}}{({\bf r}^{(k)})^T{\bf r}^{(k)}}$
            update ${\bf p}^{(k+1)}={\bf r}^{(k+1)}+\beta^{(k)}{\bf p}^{(k)}$

Test the implementation using the system created as follows:

`from sklearn.datasets import make_spd_matrix`

`np.random.seed(55555)`

`A = make_spd_matrix(6, random_state=0)`

`x_star = np.random.random(6)`

`b = np.dot(A, x_star)`




```python
# @title Solution
def CG(A, b, x0, tol=1e-5):
    xk = x0
    rk = b - np.dot(A, xk)
    pk = rk
    rk_norm = np.linalg.norm(rk)

    num_iter = 0
    vec_x = [xk]
    while rk_norm > tol:
        apk = np.dot(A, pk)
        rkrk = np.dot(rk, rk)

        alpha = rkrk / np.dot(pk, apk)
        xk = xk + alpha * pk
        rk = rk - alpha * apk
        beta = np.dot(rk, rk) / rkrk
        pk = rk + beta * pk

        num_iter += 1
        vec_x.append(xk)
        rk_norm = np.linalg.norm(rk)
        print('Iteration: {} \t x = {} \t residual = {:.4f}'.
              format(num_iter, xk, rk_norm))

    print('\nSolution: \t x = {}'.format(xk))

    return np.array(vec_x)

from sklearn.datasets import make_spd_matrix
np.random.seed(55555)
A = make_spd_matrix(15, random_state=0)
x_star = np.random.random(15)
b = np.dot(A, x_star)

print('A\n', A, '\n')
print('b\n', b, '\n')
print('The solution x* should be\n', x_star)
print('\n')

x0 = np.random.random(15)
xs = CG(A, b, x0)
```
